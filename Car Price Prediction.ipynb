## importing the datasets

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
data_train= pd.read_csv('train.csv')
data_train.head()
```

```python
data_train.isna().sum()
```

```python
data_train.info()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
def transform_labels(df,x):
    cat_cols= df.select_dtypes(include='object').columns
    le= LabelEncoder()
    for i in cat_cols:
        le.fit(x[i])
        x[i]=le.fit_transform(x[i])
```

```python
x= data_train.drop(columns='Price', axis=1)
```

```python
x.head()
```

```python
x.shape
```

```python
transform_labels(data_train,x)
```

```python
x.head()
```

```python
x.info()
```

```python
x=x.rename(columns={'Prod. year': 'Prod_year', 'Leather interior':'Leather_interior',
                    'Fuel type':'Fuel_type', 'Engine volume':'Engine_volume',
                   'Gear box type':'Gear_box_type', 'Drive wheels':'Drive_wheels'})
```

```python
x.head()
```

```python
y = data_train.Price
y
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest= train_test_split(x, y, test_size= 0.3, random_state=10)
```
